```
# 1. Electricity Bill Calculator
cust_no = int(input("Enter Customer Number: "))
units = float(input("Enter Units Consumed: "))

if units <= 100:
    bill = units * 1.5
elif units <= 300:
    bill = 100 * 1.5 + (units - 100) * 2.5
else:
    bill = 100 * 1.5 + 200 * 2.5 + (units - 300) * 4

print("Customer Number:", cust_no)
print("Bill Amount:", bill)
```

## 2. Square and Cube Divisibility

```
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

if (a**2) % b == 0 and (a**3) % b == 0:
    print("Square and cube of first number are divisible by second.")
elif (b**2) % a == 0 and (b**3) % a == 0:
    print("Square and cube of second number are divisible by first.")
else:
    print("Not divisible.")
```

## 3. Opposite Case Letters

```
a = input("Enter first letter: ")
b = input("Enter second letter: ")

if a.lower() == b.lower() and a != b:
    print("Success")
else:
    b = a.swapcase()
    print("Modified second letter:", b)
```

## 8. Adjacent Letters

```
a = input("Enter first letter: ")
b = input("Enter second letter: ")

if abs(ord(a.lower()) - ord(b.lower())) == 1:
    print("Adjacent")
else:
    print("Not Adjacent")
```

## 9. Alphabet or Special Character

```
ch = input("Enter a character: ")

if ch.isalpha():
    print("Alphabet")
else:
    print("Special Character")
```

## 10. Next and Previous Character

```
ch = input("Enter a lowercase letter: ")

next_ch = 'a' if ch == 'z' else chr(ord(ch) + 1)
prev_ch = 'z' if ch == 'a' else chr(ord(ch) - 1)

print("Previous:", prev_ch)
print("Next:", next_ch)
```

## 11. Grade Assignment

```
marks = [int(input(f"Subject {i+1}: ")) for i in range(5)]
avg = sum(marks) / 5

if avg > 35:
    print("Pass")
else:
    print("Fail")

if avg >= 85:
    print("Grade A")
elif avg >= 60:
    print("Grade B")
elif avg >= 50:
    print("Grade C")
else:
    print("Grade D")
```

## 12. Leap Year Checker

```
year = int(input("Enter year: "))

if (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0):
    print("Leap Year")
else:
    print("Not a Leap Year")
```
