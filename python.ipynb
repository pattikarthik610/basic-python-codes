```python
#star triangle pattern
n=int(input("enter a number:"))
if n>0:
  for i in range(1,n+1):
    print("*"*i)
else:
  print("please enter number greater than 0.")
```

```python
#star triangle pattern
n=int(input("enter a number:"))
if n>0:
  for i in range(n,0,-1):
    print("*"*i)
else:
  print("please enter number greater than 0.")
```

```python
#pyramid star pattern
n=int(input("enter a number:"))
if n>0:
  for i in range(1,n+1):
    spaces=" "*(n-i)
    stars="*"*(2*i-1)
    print(spaces+stars)
else:
  print("please enter number greater than 0")

```

```python
#sorting the list in ascending using sorted()
l=[7,8,6,1,0,5] #list of different elements
asc=sorted(l)
print(asc)
```

```python
#sorting the list in decending using sorted()
l=[7,8,6,1,0,5] #list of different elements
dec=sorted(l, reverse=True)
print(dec)
```

```python
# python operators 'sum','difference','product','division,'remainder','exponent.
a=int(input("enter the number:"))
b=int(input("enter the number:"))
add=a+b
diff=a-b
product=a*b
division=a/b
remainder=a%b
exponent=a**b
floordivision=a//b
print(f"Sum of a and b is:{add}.")
print(f"diff of a and b is:{diff}.")
print(f"product of a and b is:{product}.")
print(f"division of a and b is:{division}.")
print(f"remainder of a and b is:{remainder}.")
print(f"exponent of a and b is:{exponent}.")
print(f"floordivision of a and b is:{floordivision}.")
```
